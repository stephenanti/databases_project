```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('project_bbc.csv')
df
```

```python
# renaming columns to make them clearer
df = df.rename(columns={'name': 'critic', 'org':'critic_org','cn': 'critic_country'})
df
```

```python
df.groupby('critic_country')['critic'].nunique()
```

```python
new_df = df.groupby('critic_country')['movie_name'].value_counts().reset_index(name='count')
```

```python
new_df.tail(40)
```

```python
#Moving the values from the two right columns into a new column
#One that human readers can understand
new_df["string"] = new_df["movie_name"] + ": " + new_df["count"].map(str) + np.where(new_df["count"]>1, ' votes', ' vote')
new_df.tail(40)
```

```python
output = new_df.groupby('critic_country')['string'].apply(lambda x: "<div id='movie'><h1><b>Top Movies</b></h1><P>%s</P></div>" % '</p><p> '.join(x)).reset_index(name='properties.article')
# output = new_df.groupby('critic_country')['string'].apply(lambda x: "<div id='movie'><h1><b>Top Movies</b></h1><P>{0}</P></div>".format('</p><p> '.join(x))).reset_index(name='properties.article')
output
```

```python
output.iloc[10]['properties.article']
```

```python
critics = df.groupby('critic_country')['critic'].nunique().reset_index(name='properties.headline')
critics
```

```python
output = output.merge(critics, how='left', on='critic_country')
```

```python
output
```

```python
#Turn that number and into something readable.
#Though here I don't pay any attention to the English language

output['properties.headline'] = output['properties.headline'].map(str) + " critic"
```

```python
#Add some color
output['properties.color'] = "#35476E"
output
```
