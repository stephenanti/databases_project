```python
##Load the geojson file Exported from Mapshaper

with open('anti.geo.json') as json_data:
    geometry_data = json.load(json_data)
```

```python
##Normalize the hierarchy  so you have simple rows in a dataframe
##Note that you need to extract it from geometry_data['features']
geo_df = pd.DataFrame.from_dict(json_normalize(geometry_data['features']), orient='columns')
```

```python
geo_df = geo_df.rename(columns={'properties.scalerank': 'properties.scale'
                                ,'properties.featurecla':'properties.class'
                                , 'properties.labelrank':'properties.rank'
                                ,'properties.sovereignt':'properties.sov'
                                ,'properties.sov_a3':'properties.sovabb'
                                ,'properties.adm0_dif':'properties.admndiff'})
geo_df.head()
```

```python
final_output = pd.merge(output_critic, geo_df, how="left", left_on="critic_country", right_on="properties.sov")
```

```python
final_output = final_output.dropna(subset=['geometry.coordinates'])
```

```python
my_json = json.loads(final_output.to_json(orient='records'))
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geo_format = process_to_geojson(my_json)
```
