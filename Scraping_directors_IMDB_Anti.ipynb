```python
# before the loop starts, you write 
list_of_directors=[]
for count in range(0,5):
    url="https://www.imdb.com/search/keyword/?keywords=movie-director&ref_=kw_nxt&sort=moviemeter,asc&mode=detail&page="+str(count)
    raw_html=requests.get(url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    div_tags=soup_doc.find_all("div", {"class": "lister-item-content"})

    for div in div_tags:
        p_tags=div.find_all("p",{"class":"text-muted text-small"})[1]
        try: 
            dir_url = "https://www.imdb.com/"+p_tags.find("a")['href']
            list_of_directors.append(dir_url)
#             print("https://www.imdb.com/"+p_tags.find("a")['href'])
        except:
            print("No director available")
list_of_directors
```
